# Literature Review

A study by [Song, Gu & Zhang (2020)](https://www-tandfonline-com.proxy.binghamton.edu/doi/full/10.1080/13540602.2020.1719059) looked at teachers’ subjective well-being in China. They defined well-being as an umbrella term for how people perceive their own lives, circumstances, and the events happening in them. They argue that previous research that uses the concept of subjective well being encompasses life satisfaction and their experience with their emotions (Song et. al., 2020). They found that income played a significant role in teacher’s judgments of their own work, and that altruistic values were not associated with income satisfaction, but were positively related to work satisfaction (Song et. al, 2020).

An article by [Deiner & Seligman (2004)](https://pubmed-ncbi-nlm-nih-gov.proxy.binghamton.edu/26158992/), explores the concept of well-being and argues that it should be a primary concern for policymakers. This is because desirable economic and non-economic outcomes stem from by well-being and can have impacts on economic, social, and health outcomes. They argue that current economic measures do not accurately measure quality of life, and they cite basic microeconomic theory by that saying that, all else equal, having more choices leads to a higher quality of life due to the ability for a person to maximize their well-being (Deiner & Seligman, 2004). Thus, they argue that higher income should correlate with higher levels of well-being. In the past, previous research has argued that well-being rises with income until a person hits a salary of \\$75,000 dollars, at which salary stops increasing a person’s well-being. However, a new article by [Killingsworth (2021)](https://www.pnas.org/doi/10.1073/pnas.2016976118) argues that it could continue to rise after \\$75,000 dollars based on their retrospective reports. Deiner and Seligman (2004) concluded that there are 6 factors that are important for well-being: 
1. Living in a democratic, stable society
2. Having a social support system
3. Having an engaging job with stable income
4. Good physical and mental health and healthcare resources
5. Having fulfilling life goals
6. Having a guiding philosophy or religion. 

Altruism has also been largely studied. One evolutionary theory argues that altruism stems from the return from doing good things [(de Waal, 2007)](https://www-annualreviews-org.proxy.binghamton.edu/doi/10.1146/annurev.psych.59.103006.093625). Some argue that emotion acts as a motivator for helping people, as it can help the person doing the altruistic acts shift into a better mood or improve their well-being [(Barasch, Levine, Berman, & Small, 2014)](https://web-p-ebscohost-com.proxy.binghamton.edu/ehost/detail/detail?vid=0&sid=ccb3b11b-7ada-486e-adc6-ae5c162b84df%40redis&bdata=JnNpdGU9ZWhvc3QtbGl2ZQ%3d%3d#AN=97633805&db=a2h). This has led to arguments that altruism can be done for both selfish reasons (to make oneself feel good, or to relieve discomfort), or for selfless reasons (helping someone else while incurring a cost to oneself) (Barasch, Levine, Berman, & Small, 2014). Sometimes people act in altruistic ways to cooperate with their peers for group survival, although it often requires empathy (de Waal, 2007). De Waal (2007) lays out three forms of directed altruism in response to someone’s suffering. The first is altruistic impulse, which is impulsive help. The second is learned altruism, or help as a reinforced conditioned response (think classical and operant conditioning). The third is intentional altruism, which is rooted in a person’s predictions of the outcome of their help (de Waal, 2007).  

One specific study by [Rand & Epstein, (2014)](https://doi-org.proxy.binghamton.edu/10.1371/journal.pone.0109687) looked at altruism at an extreme level, at which point someone will risk their life for another. They explore different mechanisms of cooperative behavior, being direct and indirect reciprocity, population structure, and group and kin selection (Rand & Epstein, 2014). They explored a dual processing framework through which to understand decision making, arguing that one system through which people react is intuitive, fast and automatic, while the other is slow and deliberate. This led to the Social Heuristics Hypothesis (SHH), which states that the ways of being that are advantageous to us in our daily lives are internalized and become default behaviors. These behaviors can sometimes be overridden in specific situations with very deliberate processes. (Rand & Epstein, 2014). 

This study wants to further explore how different dimensions of well-being, as measured by the Sharecare well-being index, are associated with altruistic and reciprocal behavior. There is limited research on how well-being, altruism, and reciprocity are related, and it would be interesting to explore how those well-being dimensions affect those factors to dig deeper into what drives that kind of behavior. It would be interesting to see how altruism correlates with well-being. A lot of economic theory posits that people will behave in ways to maximize their own well-being. Here, well-being will be discussed in accordance with the Sharecare Well-Being index, but it should be noted that different researchers and fields of study will define and measure well-being differently. 

# Methods

Data on well-being was attained through the [Sharecare Community Well-Being Index (2020)](https://wellbeingindex.sharecare.com/), which is a combination of two indices: The well-being index and the social determinants of health index. Their [methodology](https://wellbeingindex.sharecare.com/research/community-well-being-index-methods/) can be read on their website. Their data from the [interactive map](https://wellbeingindex.sharecare.com/interactive-map/) was recreated into an excel spreadsheet for this study. This was done via estimation and comparisons of previously completed data to ensure the same values and estimation were being used consistently. Estimated values were also compared to rankings reported in their [2020 report](https://wellbeingindex.sharecare.com/download-reports/?submissionGuid=1a490b8f-54d0-48a9-bdaa-c22fd959a46b). [Global Preference Survey (GPS)](https://www.briq-institute.org/global-preferences/downloads?submitted=1) data was from the Briq institute. In their [paper](https://academic.oup.com/qje/article/133/4/1645/5025666?login=false), they stated that their data collection method was a global survey. Both data sources will be used.

The Well-Being Index describes 10 different dimensions: 
* Community: enjoys and feels safe in their community
* Economic Security: employed, insured, and financially stable
* Financial: good management over finances  
* Food Access: access to healthy food 
* Healthcare Access: access to healthcare resources 
* Housing and Transportation: stable housing value, public resources available for transport 
* Physical: in good physical health
* Purpose: enjoy day to day activities and have goals 
* Resource Access: community resources, such as libraries and churces
* Social: supportive and loving social relationships
